### Drop unnamed column from all datasets

```python
nfl2012_2014 = nfl2012_2014.iloc[:, 1:]
off2015 = off2015.iloc[:, 1:]
def2015 = def2015.iloc[:, 1:]
wl2015 = wl2015.iloc[:, 1:]
```

### Rename columns in off/def datasets

```python
offense2015 = off2015.rename(columns ={"2015 Off Tot Pts": "Off Tot/Pts", "2015 Off Pts/G": "Off Pts/G",
                                                     "2015 Off RushYds": "Off RushYds", "2015 Off RYds/G": "Off RYds/G",
                                                     "2015 Off PassYds": "Off PassYds", "2015 Off PYds/G": "Off PYds/G",
                                                     "2015 Off TotYds": "Off TotYds", "2015 Off Yds/G": "Off Yds/G"})
defense2015 = def2015.rename(columns ={"2015 Def Tot Pts": "Def Tot/Pts", "2015 Def Pts/G": "Def Pts/G",
                                                     "2015 Def RushYds": "Def RushYds", "2015 Def RYds/G": "Def RYds/G",
                                                     "2015 Def PassYds": "Def PassYds", "2015 Def PYds/G": "Def PYds/G",
                                                     "2015 Def TotYds": "Def TotYds", "2015 Def Yds/G": "Def Yds/G"}) 
```

```python
offense2015.head()
```

```python
defense2015.head()
```

### Merge Offense and Defense Data

```python
off_def2015 = pd.merge(offense2015, defense2015, on='Team')
```

```python
off_def2015.head()
```

### Rename Teams to match merge data

```python
teamnames15 = {'Arizona CardinalsArizona': 'Arizona Cardinals', 'Atlanta FalconsAtlanta': 'Atlanta Falcons', 
             'Baltimore RavensBaltimore': 'Baltimore Ravens', 'Buffalo BillsBuffalo': 'Buffalo Bills',
            'Carolina PanthersCarolina': 'Carolina Panthers', 'Chicago BearsChicago': 'Chicago Bears',
            'Cincinnati BengalsCincinnati': 'Cincinnati Bengals', 'Cleveland BrownsCleveland': 'Cleveland Browns',
            'Dallas CowboysDallas': 'Dallas Cowboys', 'Denver BroncosDenver': 'Denver Broncos', 
             'Detroit LionsDetroit': 'Detroit Lions', 'Green Bay PackersGreen Bay': 'Green Bay Packers',
            'Houston TexansHouston': 'Houston Texans', 'Indianapolis ColtsIndianapolis': 'Indianapolis Colts',
            'Jacksonville JaguarsJacksonville': 'Jacksonville Jaguars', 'Kansas City ChiefsKansas City': 'Kansas City Chiefs',
            'Miami DolphinsMiami': 'Miami Dolphins', 'Minnesota VikingsMinnesota': 'Minnesota Vikings', 
            'New England PatriotsNew England': 'New England Patriots', 'New Orleans SaintsNew Orleans': 'New Orleans Saints',
            'New York GiantsNY Giants': 'New York Giants', 'New York JetsNY Jets': 'New York Jets', 
            'Oakland RaidersOakland': 'Las Vegas Raiders', 'Philadelphia EaglesPhiladelphia': 'Philadelphia Eagles',
            'Pittsburgh SteelersPittsburgh': 'Pittsburgh Steelers', 'San Diego ChargersSan Diego': 'Los Angeles Chargers',
            'San Francisco 49ersSan Francisco': 'San Francisco 49ers', 'Seattle SeahawksSeattle': 'Seattle Seahawks',
            'St. Louis RamsSt. Louis': 'Los Angeles Rams', 
            'Tampa Bay BuccaneersTampa Bay': 'Tampa Bay Buccaneers', 'Tennessee TitansTennessee': 'Tennessee Titans',
            'Washington RedskinsWashington': 'Washington Commanders'}
off_def2015['Team'] = off_def2015['Team'].replace(teamnames15)
```

```python
off_def2015.tail(16)
```

### Put Dataset in Team Ascending Order

```python
off_def2015 = off_def2015.sort_values("Team")
```

```python
off_def2015.tail(16)
```

### Merge with Win/Loss Data

```python
nfl2015 = pd.merge(off_def2015, wl2015, on="Team")
```

```python
nfl2015.head()
```

### Merge 2012-2015 Data

```python
nfl2012_2015 = pd.concat([nfl2012_2014, nfl2015], axis=0, join='outer', ignore_index='True')
```

### Let's go a different route -bar plots

```python
team_name = 'Seattle Seahawks'
team_data = nfl2012[nfl2012['Team'] == team_name]

# Group the data by year and calculate the sum of points
points_per_year = team_data.groupby('Team')['2012 Off Tot Pts'].sum()

# Create a bar plot of points per year
points_per_year.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('2012 Off Tot Pts')
plt.title('Points per Year for {}'.format(team_name))
plt.show()
```

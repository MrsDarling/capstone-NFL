### Drop unnamed column from all datasets

```python
nfl2012_2016 = nfl2012_2016.iloc[:, 1:]
off2017 = off2017.iloc[:, 1:]
def2017 = def2017.iloc[:, 1:]
wl2017 = wl2017.iloc[:, 1:]
```

### Rename columns in off/def data

```python
offense2017 = off2017.rename(columns ={"2017 Off Tot Pts": "Off Tot/Pts", "2017 Off Pts/G": "Off Pts/G",
                                                     "2017 Off RushYds": "Off RushYds", "2017 Off RYds/G": "Off RYds/G",
                                                     "2017 Off PassYds": "Off PassYds", "2017 Off PYds/G": "Off PYds/G",
                                                     "2017 Off TotYds": "Off TotYds", "2017 Off Yds/G": "Off Yds/G"})
defense2017 = def2017.rename(columns ={"2017 Def Tot/Pts": "Def Tot/Pts", "2017 Def Pts/G": "Def Pts/G",
                                                     "2017 Def RushYds": "Def RushYds", "2017 Def RYds/G": "Def RYds/G",
                                                     "2017 Def PassYds": "Def PassYds", "2017 Def PYds/G": "Def PYds/G",
                                                     "2017 Def TotYds": "Def TotYds", "2017 Def Yds/G": "Def Yds/G"}) 
```

```python
offense2017.head()
```

```python
defense2017.head()
```

### Merge Offense and Defense Data

```python
off_def2017 = pd.merge(offense2017, defense2017, on='Team')
```

```python
off_def2017.head(32)
```

### Rename Teams to match merge data

```python
teamnames17 = {'Arizona CardinalsArizona': 'Arizona Cardinals', 'Atlanta FalconsAtlanta': 'Atlanta Falcons', 
             'Baltimore RavensBaltimore': 'Baltimore Ravens', 'Buffalo BillsBuffalo': 'Buffalo Bills',
            'Carolina PanthersCarolina': 'Carolina Panthers', 'Chicago BearsChicago': 'Chicago Bears',
            'Cincinnati BengalsCincinnati': 'Cincinnati Bengals', 'Cleveland BrownsCleveland': 'Cleveland Browns',
            'Dallas CowboysDallas': 'Dallas Cowboys', 'Denver BroncosDenver': 'Denver Broncos', 
             'Detroit LionsDetroit': 'Detroit Lions', 'Green Bay PackersGreen Bay': 'Green Bay Packers',
            'Houston TexansHouston': 'Houston Texans', 'Indianapolis ColtsIndianapolis': 'Indianapolis Colts',
            'Jacksonville JaguarsJacksonville': 'Jacksonville Jaguars', 'Kansas City ChiefsKansas City': 'Kansas City Chiefs',
            'Los Angeles ChargersLA Chargers': 'Los Angeles Chargers', 'Los Angeles RamsLA Rams': 'Los Angeles Rams',
            'Miami DolphinsMiami': 'Miami Dolphins', 'Minnesota VikingsMinnesota': 'Minnesota Vikings', 
            'New England PatriotsNew England': 'New England Patriots', 'New Orleans SaintsNew Orleans': 'New Orleans Saints',
            'New York GiantsNY Giants': 'New York Giants', 'New York JetsNY Jets': 'New York Jets', 
            'Oakland RaidersOakland': 'Las Vegas Raiders', 'Philadelphia EaglesPhiladelphia': 'Philadelphia Eagles',
            'Pittsburgh SteelersPittsburgh': 'Pittsburgh Steelers',
            'San Francisco 49ersSan Francisco': 'San Francisco 49ers', 'Seattle SeahawksSeattle': 'Seattle Seahawks', 
            'Tampa Bay BuccaneersTampa Bay': 'Tampa Bay Buccaneers', 'Tennessee TitansTennessee': 'Tennessee Titans',
            'Washington RedskinsWashington': 'Washington Commanders'}
off_def2017['Team'] = off_def2017['Team'].replace(teamnames17)
```

```python
off_def2017.tail(16)
```

### Put Dataset in Team Ascending Order

```python
off_def2017 = off_def2017.sort_values("Team")
```

```python
off_def2017.tail(16)
```

### Merge with Win/Loss Data

```python
nfl2017 = pd.merge(off_def2017, wl2017, on="Team")
```

```python
nfl2017.head()
```

### Save Dataset to Mac for future use

```python
nfl2017.to_csv('/Users/admin/Desktop/Final_Project/NFL_merged/nfl2017complete.csv')
```

### Merge 2012-2017 Data

```python
nfl2012_2017 = pd.concat([nfl2012_2016, nfl2017], axis=0, join='outer', ignore_index='True')
```

### Drop unnamed column from all datasets

```python
nfl2012_2020 = nfl2012_2020.iloc[:, 1:]
off2021 = off2021.iloc[:, 1:]
def2021 = def2021.iloc[:, 1:]
wl2021 = wl2021.iloc[:, 1:]
```

### Rename columns in off/def dataset

```python
offense2021 = off2021.rename(columns ={"Tot/Pts": "Off Tot/Pts", "2021 Off Pts/G": "Off Pts/G",
                                                     "2021 Off RushYds": "Off RushYds", "2021 Off RYds/G": "Off RYds/G",
                                                     "2021 Off PassYds": "Off PassYds", "2021 Off PYds/G": "Off PYds/G",
                                                     "2021 Off TotYds": "Off TotYds", "2021 Off Yds/G": "Off Yds/G"})
defense2021 = def2021.rename(columns ={"2021 Def Tot/Pts": "Def Tot/Pts", "2021 Def Pts/G": "Def Pts/G",
                                                     "2021 Def RushYds": "Def RushYds", "2021 Def RYds/G": "Def RYds/G",
                                                     "2021 Def PassYds": "Def PassYds", "2021 Def PYds/G": "Def PYds/G",
                                                     "2021 Def TotYds": "Def TotYds", "2021 Def Yds/G": "Def Yds/G"}) 
```

```python
offense2021.head()
```

```python
defense2021.head()
```

### Merge Offense and Defense Data

```python
off_def2021 = pd.merge(offense2021, defense2021, on='Team')
```

```python
off_def2021.head(32)
```

### Rename Teams to match merge data

```python
teamnames21 = {'Arizona CardinalsArizona': 'Arizona Cardinals', 'Atlanta FalconsAtlanta': 'Atlanta Falcons', 
             'Baltimore RavensBaltimore': 'Baltimore Ravens', 'Buffalo BillsBuffalo': 'Buffalo Bills',
            'Carolina PanthersCarolina': 'Carolina Panthers', 'Chicago BearsChicago': 'Chicago Bears',
            'Cincinnati BengalsCincinnati': 'Cincinnati Bengals', 'Cleveland BrownsCleveland': 'Cleveland Browns',
            'Dallas CowboysDallas': 'Dallas Cowboys', 'Denver BroncosDenver': 'Denver Broncos', 
             'Detroit LionsDetroit': 'Detroit Lions', 'Green Bay PackersGreen Bay': 'Green Bay Packers',
            'Houston TexansHouston': 'Houston Texans', 'Indianapolis ColtsIndianapolis': 'Indianapolis Colts',
            'Jacksonville JaguarsJacksonville': 'Jacksonville Jaguars', 'Kansas City ChiefsKansas City': 'Kansas City Chiefs',
            'Los Angeles ChargersLA Chargers': 'Los Angeles Chargers', 'Los Angeles RamsLA Rams': 'Los Angeles Rams',
            'Miami DolphinsMiami': 'Miami Dolphins', 'Minnesota VikingsMinnesota': 'Minnesota Vikings', 
            'New England PatriotsNew England': 'New England Patriots', 'New Orleans SaintsNew Orleans': 'New Orleans Saints',
            'New York GiantsNY Giants': 'New York Giants', 'New York JetsNY Jets': 'New York Jets', 
            'Las Vegas RaidersLas Vegas': 'Las Vegas Raiders', 'Philadelphia EaglesPhiladelphia': 'Philadelphia Eagles',
            'Pittsburgh SteelersPittsburgh': 'Pittsburgh Steelers',
            'San Francisco 49ersSan Francisco': 'San Francisco 49ers', 'Seattle SeahawksSeattle': 'Seattle Seahawks', 
            'Tampa Bay BuccaneersTampa Bay': 'Tampa Bay Buccaneers', 'Tennessee TitansTennessee': 'Tennessee Titans',
            'Washington Football TeamWashington': 'Washington Commanders'}
off_def2021['Team'] = off_def2021['Team'].replace(teamnames21)
```

```python
off_def2021.tail(16)
```

### Put Dataset in Team Ascending Order

```python
off_def2021 = off_def2021.sort_values("Team")
```

```python
off_def2021.tail(16)
```

### Merge with Win/Loss Data

```python
nfl2021 = pd.merge(off_def2021, wl2021, on="Team")
```

### Save Dataset to Mac for future use

### Merge 2012-2021 Data

```python
nfl2012_2021 = pd.concat([nfl2012_2020, nfl2021], axis=0, join='outer', ignore_index='True')
```
